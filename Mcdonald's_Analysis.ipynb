#### >Fats description and categories:
According to the Harvard School of Public Health, instead of promoting low-fat foods, it is important to recognize the significance of healthy fats for our overall health. Contrary to popular belief, there is generally no direct link between the percentage of calories obtained from fat and major health risks. Fats can be divided into three categories:

Unsaturated Fats: These fats are beneficial for our health and carry a lower risk of diseases. Some examples include vegetable oils, fish, nuts, and seeds.

Trans Fats: These fats are considered unhealthy and can increase the risk of diseases, even when consumed in small quantities. The category that has the highest trans fats is beef and pork(See Trans Fat Graph and Data Below).

Saturated Fats: While not as harmful as trans fats, saturated fats can still pose risks if consumed excessively. The Dietary Guidelines for Americans recommend that saturated fat intake should not exceed 10% of daily calorie intake, and the American Heart Association suggests aiming for a dietary pattern that consists of 5% to 6% of calories from saturated fats. 

```python
#Plot graph for relationship between Calories vs Calories from Fat
plt.figure(figsize=(10, 5), dpi=100) 

correlation = data['Calories'].corr(data['Calories from Fat'])
plt.scatter(data.Calories, data['Calories from Fat'], color='blue') #scatterplot for correlation
plt.text(500,450, 'r = {}'.format(round(correlation,2)))
plt.xlabel("Calories")
plt.ylabel("Calories from Fat")
plt.title("The Relationship between Calories vs Calories from Fat")
plt.show()
```

```python
data.groupby('Category')['Trans Fat'].mean()
```

```python
#Bar Graph for Percent of Trans Fat
plt.figure (figsize=(10, 5), dpi=100) 
menu_category = data.groupby('Category')['Trans Fat'].mean()
menu_category.plot.bar(color = ['red','blue','green','green','green','green','green','green','green'])
plt.title("Percent of Trans Fat")
plt.xlabel("Trans Fat")
plt.ylabel("Categories")
plt.xticks(rotation=90)
plt.show()
```

```python
data['saturated_cholesterol'] = data['Saturated Fat']/data['Cholesterol']*100
saturated_cholesterol = data.groupby('Category')['saturated_cholesterol'].mean().dropna().nlargest(5)
```

```python
#Bar Graph for Percent of Saturated Fat
plt.figure(figsize=(10, 4), dpi=100)
saturated_cholesterol.sort_values(ascending=False).plot.bar(color = 'green')
plt.title("Top 5 Categories that have the Highest Percent of Saturated Fat")
plt.ylabel("Percent")
plt.xlabel("Menu Category")
plt.xticks(rotation=90)
plt.show()
```

#### >Breakdown
Vitamin A: An essential nutrient known for its benefits in protecting the eyes, supporting the immune system, and promoting healthy growth. Among various food categories, salads have the highest concentration of Vitamin A, followed by chicken and fish, as well as smoothies and shakes.

Vitamin C: A crucial vitamin that plays a role in boosting immunity and managing high blood pressure. When it comes to McDonald's food offerings, they tend to be richer in Vitamin A compared to Vitamin C. The food categories with the highest levels of Vitamin C include salads, snacks and sides, and beverages.Within the snacks category, apple slices contain the highest amount of Vitamin C. Interestingly, McDonald's french fries also contain notable levels of Vitamin C, as potatoes are a natural source of this nutrient.

In the beverage category, it's worth noting that apple and orange juices are particularly rich in Vitamin C. However, other beverages and drinks like Coke and Sprite are generally low in vitamin content.

#### >Iron and Calcium
Shakes (fruit shakes, milk shakes etc.), coffee and tea are rich in calcium but low in iron

Menu categories rich in Animal protein (Beef & Pork, Chicken & Fish) are rich in both Iron and Calcium

Mcdonald's beverages are very low in Calcium and Iron.This analysis presents a comprehensive overview of the McDonald's Menu, highlighting the nutritional value of each food category. With this information, you can make informed choices based on your personal preferences or specific nutritional requirements.

McDonald's offers a variety of foods that are abundant in essential vitamins, particularly Vitamin A and C, with salads standing out as a particularly rich source. Additionally, many menu items are low in carbohydrates and contain little to no trans-fat, which is a positive aspect. However, it is important to note that several foods are high in saturated fat. If you are watching your fat intake, it is advisable to select options with low saturated fat content, aiming for less than 10%.

```python
iron = data.groupby('Category')['Iron (% Daily Value)'].mean()
calcium = data.groupby('Category')['Calcium (% Daily Value)'].mean()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

category = iron.index
Iron = iron.values
Calcium = calcium.values

X_axis = np.arange(len(category))

plt.bar(X_axis - 0.2, Iron, 0.4, label = 'Iron', color = 'green')
plt.bar(X_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'yellow')

plt.xticks(X_axis, category)
plt.xlabel("Menu Categories")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
